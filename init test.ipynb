```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('init.csv')
from IPython import display
```

```python
TEST = []
for i in range(1,14):
    TEST.append(data[data['testSet']=="Test{:02}".format(i)])
for i in range(1,13):
    print("Test{:02}".format(i))
    print(TEST[i].mean())
```

```python
Set = "Test13"
TEST1 = data[data['testSet']==Set]
plt.plot(TEST1.groupby('sec', as_index=True)['intermidiary'].mean(),c='b')
plt.plot(TEST1.groupby('sec', as_index=True)['final'].mean(),c='r')
plt.legend()
plt.title("Test Set {}".format(Set[-2:]))
plt.xlabel("Seconds")
plt.ylabel("Objective value")
plt.savefig("/Users/jens/Dropbox/Apps/ShareLaTeX/Metaheuristic/Figures/{}.png".format(Set))
plt.show()
```
